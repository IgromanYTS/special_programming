Імпортуємо модулі

```python
import pandas as pd
import numpy as np
import timeit
```

Імпортуємо датасет через pandas

```python
data = pd.read_csv("household_power_consumption.txt", sep=";", na_values="?")
data
```

Робимо перше завдання через pandas

```python
def global_active(k):
    return data[data["Global_active_power"] > k]
    
```

Робимо друге завдання через pandas

```python
def voltage(l):
    return data[data["Voltage"]>l]
```

Робимо третє завдання через pandas

```python
def intensity_task3(intensity1, intensity2):
    return data[((data["Global_intensity"] > intensity1) & (data["Global_intensity"] < intensity2)) & (data["Sub_metering_2"] > data["Sub_metering_3"])]
```

Робимо четверте завдання через pandas

```python
def func4(a):
    data4 = data.sample(n=a, replace=True)
    mean1 = data4["Sub_metering_1"].mean()
    mean2 = data4["Sub_metering_2"].mean()
    mean3 = data4["Sub_metering_3"].mean()
    return (f"Sub_metering_1: {mean1}, Sub_metering_2: {mean2}, Sub_metering_3: {mean3}")
```

```python
func4(500000)
```

Робимо пʼяте завдання через pandas

```python
def func5():
    data5 = data[ (data["Time"] > "18:00:00") & (data["Global_active_power"] > 6) & (data["Sub_metering_2"] > data["Sub_metering_1"]) & (data["Sub_metering_2"] > data["Sub_metering_3"])]

    half1 = data5.iloc[:len(data5) //2]
    half2 = data5.iloc[len(data5)//2:]
    select_from1 = half1.iloc[::3]
    select_from2 = half2.iloc[::4]
    result = pd.concat([select_from1, select_from2])
    return result

func5()
```

Імпортуємо датасет через numpy

```python
data_np = np.genfromtxt("household_power_consumption.txt", delimiter=";", missing_values="?", dtype=str, skip_header=1, filling_values=0)
data_np[data_np == "?"] = np.nan
data_np
```

Робимо перше завдання через numpy

```python
def task1(j):
    current = data_np[:, 2].astype(float)
    return current[current > j]
```

```python
task1(5)
```

Робимо друге завдання через numpy

```python
def task2(l):
    current = data_np[:, 4].astype(float)
    return current[current > l]
```

```python
task2(235)
```

Робимо третє завдання через numpy

```python
def task3(n, k):
    current = data_np[:, 5].astype(float)
    condition1 = data_np[:, 7].astype(float)
    condition2 = data_np[:, 8].astype(float)
    return current[((current > n) & (current < k) ) & (condition1 > condition2)]
```

```python
def task3_1():
    current = data_np[:, 5].astype(float)
    condition1 = data_np[:, 7].astype(float)
    condition2 = data_np[:, 8].astype(float)
    return current[((current > 19) & (current < 20) ) & (condition1 > condition2)]
```

```python
task3(19, 20)
```

Робимо четверте завдання через numpy

```python
def task4(a):
    n_arr = np.random.choice(len(data_np),a ,replace=True)
    array = data_np[n_arr]
    #return array
    mean1 = np.nanmean(array[:,6].astype(float))
    mean2 = np.nanmean(array[:,7].astype(float))
    mean3 = np.nanmean(array[:,8].astype(float))
    return (f"Mean1: {mean1}, Mean2: {mean2}, Mean3: {mean3}")
```

```python
task4(500000)
```

Робимо пʼяте завдання через numpy

```python
def task5():
    current_time = data_np[:, 1]
    current_power = data_np[:, 2].astype(float)
    condition1 = data_np[:, 6].astype(float)
    condition2 = data_np[:, 7].astype(float)
    condition3 = data_np[:, 8].astype(float)
    data_new = data_np[ (current_time > "18:00:00") & (current_power > 6) & (condition2 > condition1) & (condition2 > condition3) ]
    
    count = len(data_new) // 2
    data_1 = data_new[:count]
    data_2 = data_new[count:]
    return np.concatenate([data_1[::3], data_2[::4]])
```

```python
task5()
```

```python
def time_mesure(func, arg):
    if arg == 0:
        time_f = timeit.timeit(lambda: func(), setup="pass", number=1)
    else: 
        time_f = timeit.timeit(lambda: func(arg), setup="pass", number=1)
    print(f"time: {time_f:.3f} seconds")
    print("-"* 20)
```

Робимо заміри часу для pandas

```python
print("PANDAS")
time_mesure(global_active, 5)
time_mesure(voltage, 235)
time_mesure(intensity_task3_1, 0)
time_mesure(func4, 500000)
time_mesure(func5, 0)
```

Робимо заміри часу для numpy

```python
print("NUMPY")
time_mesure(task1, 5)
time_mesure(task2, 235)
time_mesure(task3_1, 0)
time_mesure(task4, 500000)
time_mesure(task5, 0)
```
